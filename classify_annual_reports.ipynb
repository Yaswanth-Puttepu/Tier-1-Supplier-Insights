```python
import pandas as pd
from collections import Counter

# Load your CSV file into a pandas DataFrame
df = pd.read_csv('list_of_urls.csv')

# Define the classification function with handling for non-string values
def classify_url(url):
    if isinstance(url, str):  # Check if the value is a string
        url_lower = url.lower()
        if url_lower == 'n':
            return 'No Annual Report'
        elif 'about' in url_lower:
            return 'About Page'
        elif 'investor' in url_lower or 'ir' in url_lower or 'inversores' in url_lower:
            return 'Investors Page'
        elif 'news' in url_lower or 'press' in url_lower:
            return 'News Section'
        else:
            return 'Other'
    else:
        return 'Other'  # Default category for non-string values

# Apply the classification function to the DataFrame
df['Category'] = df['links'].apply(classify_url)  # Use the correct column name here

# Separate the DataFrame into different categories
about_page_df = df[df['Category'] == 'About Page']
investors_page_df = df[df['Category'] == 'Investors Page']
news_section_df = df[df['Category'] == 'News Section']
no_annual_report_df = df[df['Category'] == 'No Annual Report']
other_page_df = df[df['Category'] == 'Other']

# Save each category to a separate CSV file
about_page_df.to_csv('about_page_annual_reports55.csv', index=False)
investors_page_df.to_csv('investors_page_annual_reports55.csv', index=False)
news_section_df.to_csv('news_section_annual_reports55.csv', index=False)
no_annual_report_df.to_csv('no_annual_report55.csv', index=False)
other_page_df.to_csv('other_page_annual_reports55.csv', index=False)

# Print the count of each category
print(f"About Page URLs: {len(about_page_df)}")
print(f"Investors Page URLs: {len(investors_page_df)}")
print(f"News Section URLs: {len(news_section_df)}")
print(f"No Annual Report URLs: {len(no_annual_report_df)}")
print(f"Other Page URLs: {len(other_page_df)}")

# Optionally, print the first few rows of each category to verify
print("\nAbout Page URLs:")
print(about_page_df.head())

print("\nInvestors Page URLs:")
print(investors_page_df.head())

print("\nNews Section URLs:")
print(news_section_df.head())

print("\nNo Annual Report URLs:")
print(no_annual_report_df.head())

print("\nOther Page URLs:")
print(other_page_df.head())
```
